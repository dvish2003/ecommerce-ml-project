### Download dataset

```python
from datasets import load_dataset
import pandas as pd
import os

dataset = load_dataset("millat/e-commerce-orders")
df = dataset["train"].to_pandas()

# create root data folder
os.makedirs("../data/", exist_ok=True)

# save into root data folder
df.to_csv("../data/ecommerce_orders.csv", index=False)
```
